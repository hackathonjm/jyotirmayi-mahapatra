```python
import numpy as np
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['Mall_Customers.csv']))
df.shape
```

```python
df.head()
```

```python
df["A"]= df["Annual Income (k$)"]
df["B"]= df["Spending Score (1-100)"]
```

```python
x=df[["A","B"]]
x.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
plt.scatter(x["A"],x["B"], s = 30, c = 'b')
plt.show()
```

```python
kmean = KMeans(n_clusters=5)
kmean.fit(x)
centers = kmean.cluster_centers_
print(kmean.cluster_centers_)
```

```python
clusters = kmean.fit_predict(x)
df["label"] = clusters
df.head(100)
```

```python
col=['green', 'blue', 'red', 'yellow', 'magenta']
```

```python
for i in range(5):
  a=col[i]
  plt.scatter(df.A[df.label == i],df.B[df.label == i], c = a, label = f'cluster 1')
  plt.scatter(centers[:,0], centers[:,1], marker='*', s = 300, c='r', label='centroid')
```

```python
x1 = x.loc[:,["A","B"]].values
```

```python
wcss = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k, init = "k-means++")
  kmeans.fit(x1)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```
